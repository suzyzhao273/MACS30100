```python
import pandas as pd

# 1. read current final table
df = pd.read_csv("final.csv")
df.columns = df.columns.str.strip()
df["artist_id"] = df["artist_id"].astype(str).str.strip()

print("Original final shape:", df.shape)
print("Original columns:", df.columns.tolist())


# 2. read artist metadata
artists = pd.read_csv("musicoset_metadata/artists.csv", sep="\t", low_memory=False)
artists.columns = artists.columns.str.strip()
artists["artist_id"] = artists["artist_id"].astype(str).str.strip()

print("\nArtists raw shape:", artists.shape)
print("Artists columns:", artists.columns.tolist())


# 3. keep only needed columns
artist_controls = artists[["artist_id", "followers", "artist_type", "main_genre"]].copy()
artist_controls = artist_controls.drop_duplicates(subset=["artist_id"])

print("\nArtist controls shape after dedup:", artist_controls.shape)


# 4. merge into final table
df2 = df.merge(artist_controls, on="artist_id", how="left")

print("\nMerged final shape:", df2.shape)


# 5. check coverage
for col in ["followers", "artist_type", "main_genre"]:
    print(f"{col} coverage:", f"{(1 - df2[col].isna().mean()):.2%}")


# 6. preview new columns
print("\nPreview:")
print(df2[["artist_id", "followers", "artist_type", "main_genre"]].head())


# 7. save as a NEW file
df2.to_csv("final_with_artist_controls.csv", index=False)

print("\nSaved to: final_with_artist_controls.csv")
```
